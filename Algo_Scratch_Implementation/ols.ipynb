# Ordinary Lesat Square (OLS)

```python
import pandas as pd
import numpy as np
```

```python
class LinearRegressionOLS:
    def __init__(self):
        self.bias=None
        self.weights=None
    def fit(self,X,y):
        m,n=X.shape
        X_b=np.c_[np.ones((m,1)),X]
        #Normal Equation
        # theta=np.dot(np.linalg.inv(np.dot(X_b.T,X_b)),np.dot(X_b.T,y))
        theta=np.linalg.inv(X_b.T @ X_b)@ X_b.T @ y # @ here defines the dot product

        self.bias=theta[0]
        self.weights=theta[1:]
    def predict (self,X):
        y_pred=self.bias+np.dot(X,self.weights)
        return y_pred
        
```

```python
X=np.array([[1],[2],[3],[4],[5]])
y=np.array([2,4,6,8,10])
model=LinearRegressionOLS()
model.fit(X,y)
y_pred=model.predict(X)
print(y_pred)
print(model.bias)
print(model.weights)
```
