# Scratch implementaion from KNN Algorithm

```python
import pandas as pd
import numpy as np
```

```python
class KNNClassifier:
    def __init__(self,k=3):
        self.k=k
    def _euclidean_dist(self,x1,x2):
        return np.sqrt(np.sum((x1-x2)**2))
    def fit(self,X,y): #Lazy Learner
        self.X_train=X
        self.y_train=y
    def _predict_one(self,x):
        distance=[self._euclidean_dist(x,x_train) for x_train in self.X_train]
        knn_indices=np.argsort(distance)[:self.k]
        knn_classes=[self.y_train[i] for i in knn_indices]
        majority_class=np.argmax(np.bincount(knn_classes))
        return majority_class
    def predict (self,X):
        y_pred=[self._predict_one(X) for x in X]
        return np.array(y_pred)
        
```

```python
# Data
X_train=np.array([
    [1,2],[2,3],[3,3],[4,5],[5,6],[7,7]
])
y_train=np.array([0,0,0,1,1])
X_test=np.array([
    [2,2],
    [6,6]
])

model=KNNClassifier()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(y_pred)
```
