# Max Doctors Lists

```python
import requests 
import pandas as pd
from bs4 import BeautifulSoup
```

```python
with open("max_doc_data.html","r")as f:
    data=f.read()
soup=BeautifulSoup(data,"lxml")
name=soup.select("div.doctor-card-new-text")
max_doc=[]
for d in name:
    # print(d)
    n=d.select_one("h4 a")
    post = d.select_one("p")
    spec=d.select_one("p.text-truncate-line-3")
    experience = ""
    gender = ""
    all_p = d.select("p")

    for p in all_p:
        text = p.text.strip()

        if "Experience" in text:
            experience = text.replace("Experience:", "").strip()

        if "Gender" in text:
            gender = text.replace("Gender:", "").strip()

            
    
    name=n.get_text(strip=True)
    post=post.get_text(strip=True)
    specialist=spec.get_text(strip=True)
    max_doc.append([name or None,experience or None,gender or None,specialist or None,post or None])
doctors=pd.DataFrame(max_doc,columns=["Name","Total-Experience","Gender","Specialist","Post"])
doctors.to_csv("max_doctors.csv",index=False)

doctors
    
```
