```python
import requests
from bs4 import BeautifulSoup
```

```python
page_count=1
while True:
    url=f"http://quotes.toscrape.com/page/{page_count}/"
    res=requests.get(url)
    soup=BeautifulSoup(res.text,"lxml")
    quotes=soup.select("div.quote")

    if not quotes:
        print("No valid page anymore")
        break
    with open(f"quotes_data_page{page_count}.html","w",encoding="utf-8")as f:
        f.write(res.text)
        print(f"downloaded data from {page_count}")
    page_count+=1

```
