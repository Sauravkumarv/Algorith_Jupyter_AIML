# Scrapping data with Beautifull Soup

```python
from bs4 import BeautifulSoup
import pandas as pd
with open("scrap_data.csv","r") as f:
    html_content=f.read()

soup=BeautifulSoup(html_content,"lxml")
all_tr=soup.find_all("tr",class_="team")
all_tr

h_teams=[]
for tr in all_tr:
    t1=tr.find("td",class_="name")
    y=tr.find("td",class_="year")
    w=tr.find("td",class_="wins")
    l=tr.find("td",class_="losses")
    
    team=t1.get_text(strip=True)
    year=y.get_text(strip=True)
    win=w.get_text(strip=True)
    loss=l.get_text(strip=True)

    h_teams.append([team,year,win,loss])

Teams=pd.DataFrame(h_teams,columns=["Team","Year","Win","Loss"])
Teams
# Teams.to_csv("normalize_data.csv",index=False)
```
