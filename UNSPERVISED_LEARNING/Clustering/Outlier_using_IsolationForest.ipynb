# Outlier/Anomalies Detection using Isloation Forest

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
```

```python
df=pd.read_csv("thyroid_dataset.csv")
X=df.drop("Outlier_label",axis=1)
y=df["Outlier_label"]
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
clf=IsolationForest(
    n_estimators=200,
    contamination='auto',
    random_state=42,
    )
label=clf.fit_predict(X_scaled)
```

# Dimesionality Reduction using PCA (Principle Component Analysis)

```python
from sklearn.decomposition import PCA
import numpy as np
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0],X_pca[:,1],c=label)
plt.xlabel("PC1")
plt.ylabel("PC2")
n_outliers=np.sum(label==-1)
n_normal=np.sum(label==1)
print(n_outliers)
print(n_normal)
```

```python
from sklearn.neighbors import LocalOutlierFactor
neighbors=LocalOutlierFactor()
labels=neighbors.fit_predict(X_scaled)
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0],X_pca[:,1],c=labels)
plt.xlabel("PC1")
plt.ylabel("PC2")
n_outliers=np.sum(labels==-1)
n_normal=np.sum(labels==1)
print(n_outliers)
print(n_normal)
```
