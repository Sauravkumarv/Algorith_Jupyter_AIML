# Principal Comopenet Analysis (PCA)

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
iris=load_iris()
X=iris.data
y=iris.target
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)
X_pca.shape
print("PCA expalined variance ratio",pca.explained_variance_ratio_) ## yha iska matlb hai kitna percent of data pe ye dimensionality reduction apply hua hai
```

```python
plt.scatter(X_pca[:,0],X_pca[:,1],c=y)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA for our Dataset")
```
