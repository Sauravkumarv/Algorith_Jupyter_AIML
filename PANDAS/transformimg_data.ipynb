# Applying Functions

```python
import pandas as pd
df=pd.read_csv("Practice/raw_data.csv")
df1=df.copy()
df1["tax"]=df1["income"].apply(lambda x:"20%" if x>50000 else "10%")
df1
```

```python
df1["gender"]=df1["gender"].map({"Male":'M',"Female":'F'})
df1
```

```python
df2=df1.assign(new_income=df["income"]*1.5)
df2
```

```python
df2["country"]=df2["country"].replace({"USA":"us","Canada":"cd","Spain":"sp","China":"ch","India":"in","Mexico":"mx"})
df2.fillna(0)
```

```python
df3=df2.rename( columns={"new_income":"after_appraisel"})
df3
```
