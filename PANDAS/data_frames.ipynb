# A DataFrame is a two-dimensional, tabular data structure (like a spreadsheet or SQL table)

```python
import pandas as pd
info={
    "Name":["Adam","Eve","Bob"],
    "Age":[78,99,85],
    "Grade":["A","B","O"],
}
arr=pd.DataFrame(info)
print(arr)
type(arr)
```

```python
print(arr.index)
print(arr.columns)
```

```python
import numpy as np
np_arr=np.array([[1,4,5],[2,3,6],[9,8,7]])
pd_arr=pd.DataFrame((np_arr),columns=["col1","col2","col3"])
print(pd_arr)
```

```python
l=["Adam", 96], ["Eve", 75], ["Bob", 82], ["Charlie", 92]
df=pd.DataFrame(l,columns=["Name","Marks"])
print(df)
```
