# Basic of Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
taxis=sns.load_dataset("tips")
taxis.head()
```

```python
sns.set_theme(style="darkgrid")
sns.relplot(
    data=taxis,
    x="total_bill",
    y="tip",
    size="tip",
   
)
plt.title("Avearge Tipped amount")
```

```python
sns.relplot(
    data=taxis,
    y="tip",
    x="total_bill",
    hue="sex",
    style="time",
)
```

```python
sns.relplot(
    data=taxis,
    y="tip",
    x="total_bill",
    hue="sex",
    style="time",
)
```

```python
sns.get_dataset_names()
flights=sns.load_dataset("flights")
sns.set_theme(style="white")
sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)
```

```python
sns.barplot(
    data=taxis,
    x="day",
    y="tip",
    hue="sex"
    
)
```

```python
sns.boxplot(
    data=taxis,
    x="day",
    y="tip",
    hue="sex"
    
)
```

```python
flights.head()
```

```python
fight=flights.pivot(index="month",columns="year",values="passengers")
```

```python
sns.heatmap(
    data=fight,
    cmap="cool",
    annot=True,
    fmt="d"
   
)
plt.title("Passenger Heatmap")
plt.tight_layout()
```
