# Scatter plot of total bill

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
totalbill=tips["total_bill"]
tip=tips["tip"]
sex=tips["sex"]
smoker=tips["smoker"]


fig,ax=plt.subplots()
fig.suptitle("Total Bill vs Tip")
colors = ["red" if s == "Male" else "green" for s in sex ]

ax.scatter(
    totalbill,
    tip,
    color=colors,
    
)
plt.xlabel("total_bill in ($)")
plt.ylabel("tips")
```

```python
sns.set_theme(style="darkgrid")
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker"
)
plt.title("Total Bill vs Tip")
```
