# Seaborn penguine datasets

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
peguine=sns.load_dataset("penguins")
body_mass_g=peguine["body_mass_g"]
flipper_length_mm=peguine["flipper_length_mm"]
peguine.head()
fig,ax=plt.subplots()
fig.suptitle("Penguine bodymass by flipper length")
fig.supxlabel("body_mass_g")
fig.supylabel("flipper_length_mm")
ax.scatter(body_mass_g,flipper_length_mm)

```

```python
sns.set_theme(style="darkgrid")
sns.relplot(
    data=peguine,
    x="body_mass_g",
    y="flipper_length_mm",
    hue="species"
)
plt.title("Penguine bodymass by flipper length")
plt.tight_layout()
```
