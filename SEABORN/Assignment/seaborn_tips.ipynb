# Seaborn Tips Datasets

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")
tips.sample()
```

```python
tips1=tips.pivot_table(index="day",columns="time",values="tip",observed=True)
```

```python
tips1
sns.heatmap(
    data=tips1,
    annot=True,
    cmap="crest"
    )
plt.title("Best performance day")
```
