# Histogram of Scores

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

scores=np.random.normal(loc=70,scale=10,size=200)
fig,ax=plt.subplots()  ### Modern matpltlib
fig.suptitle("students scores")
fig.supxlabel("scores")
fig.supylabel("Students counts")

ax.hist(
    scores,bins=20,
    edgecolor="black",
    color="green"
)
ax.axvline(scores.mean(),color="red")
fig.tight_layout()

## Most students are scored nearby average
```

```python
sns.set_theme(style="darkgrid")
sns.histplot(
    data=scores,
    bins=20,
    
)
plt.title("students scores")

plt.axvline(scores.mean(),color="green",linestyle="--",linewidth=2)
```
