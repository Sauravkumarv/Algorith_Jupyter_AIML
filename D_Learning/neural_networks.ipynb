# Starting Deep learning with percetron

```python
import torch
import torch.nn as nn
```

```python
# Input
inputs=torch.tensor([1.0,2.0,3.0])
neuron=nn.Linear(in_features=3,out_features=1)
```

```python
# Find weigth
neuron.weight
neuron.bias
```

```python
output=neuron(inputs)
output
```

```python
# Verify .... formula of weighted sum
neuron.weight@inputs+neuron.bias
```
