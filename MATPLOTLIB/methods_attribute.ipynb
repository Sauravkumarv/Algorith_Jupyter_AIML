# Methods and attribute in matplotlib

```python
import matplotlib.pyplot as plt
products = ["Laptop", "Mobile", "Headphones", "Smartwatch"]

online_sales  = [150, 500, 180, 820]
offline_sales = [100, 250, 140, 90]
plt.plot(products,online_sales,label="Online Sales",marker=".",linestyle="-.",color="yellow")
plt.plot(products,offline_sales,label="Offline Sales",marker="*",linestyle=":",color="green",linewidth="4")
plt.xlabel("Device")
plt.grid(False)
plt.style.use("dark_background")
plt.ylabel("Sales")
plt.legend()
plt
```
