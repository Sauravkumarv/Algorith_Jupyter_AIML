```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
score=np.random.normal(60,10,100)
bins=[0,30,70,90]
plt.hist(score,bins=bins,color="skyblue",edgecolor="black")
plt.xlabel("scores")
plt.ylabel("# students marks")
```

```python
marks = [30, 45, 48, 50, 55, 57, 60, 61, 63, 67, 70, 72, 75, 78, 80, 85, 90]

plt.hist(marks,edgecolor="black",bins=20)
plt.xlabel("marks")
plt.ylabel("No. of students")
```

```python
salaries = [15, 20,25,25, 25,30,30, 22]
plt.hist(salaries,edgecolor="black",bins=10)
plt.xlabel("salaries")
plt.ylabel("workers")
```

# Multi Histogram 

```python
legit_transactions = [
    2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
    45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99,
    249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]

fraud_transactions = [
    50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
    1200, 1500, 2000, 2500, 3000, 8000
]


plt.hist(fraud_transactions,edgecolor="black",alpha=0.6,color="red",label="fraud_transactions")
plt.hist(legit_transactions,edgecolor="black",color="green",bins=5,label="legit_transactions")
plt.grid(True)
plt.title("Fraud Detection distributed chart")
plt.xlabel("Amount in($)")
plt.ylabel("Frequency")
plt.axvline(1500,label="Maximum amount need to fraud",linewidth=2,linestyle="--")
plt.legend()
```
