# Practising Pie Chart

```python
import matplotlib.pyplot as plt
labels = ["Rent", "Food", "Transport", "Shopping", "Savings"]
values = [40, 20, 10, 15, 15]
explode = ([0, 0, 0, 0.1, 0])

plt.figure(figsize=(8,8))
plt.title("Monthly Expenses")
plt.pie(values,labels=labels,autopct="%1.1f%%",radius=1,startangle=90,explode=explode,
       wedgeprops = {
    "edgecolor": "black",   # border color
    "linewidth": 1.5,       # border thickness
    "linestyle": "--",      # dashed border
    "alpha": 0.9            # transparency
}
       )
plt.legend(loc="upper right",fontsize=8)

plt.savefig("pieChart.png")

```
