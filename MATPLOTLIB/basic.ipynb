# Basic of Matplotlib

```python
import matplotlib.pyplot as plt
students = ["Aman", "Riya", "Karan", "Neha", "Rohit", "Priya"]
maths_marks = [78, 85, 69, 92, 74, 88]
science_marks = [72, 90, 65, 95, 70, 85]

plt.plot(students,maths_marks,label='Math_marks',color="orange",marker="*",linestyle="-.",linewidth="3")
plt.plot(students,science_marks ,label='Science_marks',color="green",marker="+",linestyle=":")
plt.title("Students and their marks")
plt.xlabel("Students Name")
plt.ylabel("Marks")
plt.legend()
```
