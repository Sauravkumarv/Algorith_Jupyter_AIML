# Types of Charts

```python
import matplotlib.pyplot as plt
```

# Line Plot Chart

```python
epochs = list(range(1, 11))

train_loss = [0.95, 0.85, 0.78, 0.70, 0.65, 0.60, 0.56, 0.53, 0.51, 0.49]
val_loss   = [1.05, 0.92, 0.85, 0.80, 0.78, 0.76, 0.75, 0.74, 0.73, 0.72]

plt.plot(epochs,train_loss,label="Train_loss",marker=".")
plt.plot(epochs,val_loss,label="Validation_loss",marker="*")
plt.title("Training Loss")
plt.xlabel("Epochs")
plt.ylabel("Train & Validations")
plt.legend()
plt.xlim(0,10.8)
plt.ylim(0.5,1.2)
for i in range (len(train_loss)):
    plt.annotate(
        f"{train_loss[i]}",
        xy=(epochs[i], train_loss[i]),
        xytext=(epochs[i], train_loss[i]+0.01 )
        
    )
for i in range (len(val_loss)):
    plt.annotate(
        f"{val_loss[i]}",
        xy=(epochs[i], val_loss[i]),
        xytext=(epochs[i], val_loss[i]+0.01 )
        
    )
```
