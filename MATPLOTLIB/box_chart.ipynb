# Box_chart

```python
import matplotlib.pyplot as plt
marks = [35, 40, 45, 50, 55, 60, 65, 70, 75, 80, 90]

plt.boxplot(marks,label="new")
plt.title("Distibution Of DATA Sets")
plt.xlabel("Marks")
plt.ylabel("Distibution")
plt.legend()
```

# Multiple Box_Plots

```python
data1=[7,8,5,6,9,7,8,10,4,6]
data2=[5,6,7,8,2,5,4,6,7,5, 6] 

plt.boxplot([data1,data2],tick_labels=["data1","data2"],meanline=True,showmeans=True)
```
