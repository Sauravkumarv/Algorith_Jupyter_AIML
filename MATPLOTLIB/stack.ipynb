# Stack Plot

```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5]
y1=[5,2,3,4,5]
y2=[2,1,2,1,2]

plt.stackplot(x,y1,y2,labels=['Data1', 'Data2'], colors=['blue', 'green'])
plt.xlabel("X")  
plt.ylabel("Y") 
plt.title("Basic Stack Plot") 
plt.legend(loc='upper left')
plt.show()
```

```python
years = [2018, 2019, 2020, 2021, 2022]

children = [20, 22, 24, 25, 27]
adults   = [50, 52, 55, 57, 60]
seniors  = [10, 11, 12, 13, 14]

plt.stackplot(years,children,adults,seniors,labels=["children","adults","seniors"])
plt.xlabel("years")
plt.ylabel("Population (in lakhs)")
plt.title("City Population Growth - Stack Plot")
plt.xticks(years) 
plt.legend()
plt.tight_layout()
```
