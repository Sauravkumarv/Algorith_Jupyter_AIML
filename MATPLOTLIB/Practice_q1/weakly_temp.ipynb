# Weekly Temperature

```python
import matplotlib.pyplot as plt
import numpy as np
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
    [22, 23, 21, 24, 25],  # New York
    [18, 19, 17, 20, 21],  # London
    [30, 32, 31, 33, 34],  # Delhi
    [25, 26, 24, 27, 28]   # Tokyo
]

fig,ax=plt.subplots(2,2)
fig.suptitle("Weekly tempearture of mention cities")

# using loops
t=0
for i in range(2):
    for j in range(2):
        ax[i][j].plot(days,temperatures[0],marker="o")
        ax[i][j].set_title(cities[t])
        ax[i][j].set_xlabel
        ax[i][j].grid(True)
        
        t+=1
fig.supylabel("Temperature(celcius)")
fig.supxlabel("-- Weeks --")
fig.tight_layout()
plt.show()
    



```
