# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
cities = ["Delhi", "Mumbai", "Bangalore", "Chennai"]

male_population   = [9.8, 7.5, 6.2, 4.3]
female_population = [8.9, 7.0, 5.8, 4.0]
x=np.arange(len(cities))

width=0.4
plt.bar(x-width/2,male_population,width, label="Male")
plt.bar(x+width/2,female_population,width, label="Female")
plt.title("Population in popular cities")
plt.xlabel("Cities")
plt.ylabel("Population in (cr)")
plt.legend()
plt.xticks(x,cities)

for m in range(len(male_population)):
    plt.text(
        x[m]- width/2,
        male_population[m]+0.1,
        male_population[m],
        ha='center',
            )
    plt.text(
        x[m]+ width/2,
        female_population[m]+0.1,
        female_population[m],
        ha='center',
            )
```

# Horizontal Chart

```python
import random

categories = ["A", "B", "C", "D", "E"]
data_1 = [random.randint(10, 120) for _ in categories]
data_2 = [random.randint(10, 120) for _ in categories]

x=np.arange(len(data_1))
heigth=0.4

plt.title("Random Bar Data Generator")
plt.ylabel("Categories")
plt.xlabel("Data")
plt.barh(x-heigth/2,data_1,heigth,label="Data_1",color="green")
plt.barh(x+heigth/2,data_2,heigth,label="Data_2",color="black")
plt.yticks(x,categories)
plt.legend()
```
