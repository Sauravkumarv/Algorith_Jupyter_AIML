# Scatter Chart

```python
import matplotlib.pyplot as plt
Name= ["Amit", "Rahul", "Neha", "Priya", "Suresh", "Anita", "Rohit", "Kavita"]
Age=[25, 30, 35, 40, 45, 50, 55, 60]
BP= [110, 115, 120, 130, 135, 140, 150, 160]
colors=['green' if bp <135 else 'red' for bp in BP ]

plt.title("Age with Blood Presser")
plt.xlabel("Age")
plt.ylabel("Blood Presser")
plt.scatter(Age,BP,s=BP,cmap="cool",c=BP)
plt.colorbar(label="BP")
plt.ylim(100,170)
plt.xlim(20,70)
for i in range(len(Name)):
    plt.annotate(
        Name[i],
        xy=(Age[i],BP[i]),
        xytext=(Age[i]-4,BP[i]),
    )
```
