```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score,accuracy_score,confusion_matrix,classification_report,recall_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier,VotingClassifier
```

```python
df=pd.read_csv("novagen_dataset.csv")
en=["Diet_Type__Vegan","Diet_Type__Vegetarian","Blood_Group_AB","Blood_Group_B","Blood_Group_O"]
```

```python
ohe=OneHotEncoder(drop="first",sparse_output=False,handle_unknown="ignore")
df[en]=ohe.fit_transform(df[en])

X=df.drop("Target",axis=1)
y=df["Target"]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42,stratify=y)
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
svc=SVC()
svc.fit(X_train_scaled,y_train)
```

```python
y_pred=svc.predict(X_test_scaled)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(precision_score(y_test,y_pred))
```

 # Now Using Pipeline and cross validation 

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

pipeline=Pipeline([
    ("scaler",StandardScaler()),
    ("svc",SVC())
])
param_grid = {
    "svc__C": [0.1, 1, 10],
    "svc__kernel": ["linear", "rbf"]
}
clf = GridSearchCV(
    pipeline,
    param_grid,
    cv=5,
    scoring="recall"
)
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(precision_score(y_test,y_pred))
```

# Now Best Way to find

# Creating a Base Line Model

```python
X=df.drop("Target",axis=1)
y=df["Target"]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42,stratify=y)
result={}
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

# LOGISTCS REGRESSION (L1)

```python
log_reg=LogisticRegression(penalty="l2",
    solver="liblinear",
    max_iter=1000)
log_reg.fit(X_train_scaled,y_train)
y_pred=log_reg.predict(X_test_scaled)

# In Model Evaluation, Recall is more important than accuracy 
# because missing a high-risk patient is dangerous
result["Logistics Regression"]=recall_score(y_test,y_pred)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(recall_score(y_test,y_pred))
```

# KNN

```python
knn=KNeighborsClassifier(n_neighbors=5,metric="euclidean")
knn.fit(X_train_scaled,y_train)
y_pred=knn.predict(X_test_scaled)

result["KNN"]=recall_score(y_test,y_pred)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(recall_score(y_test,y_pred))
```

# RANDOM FOREST

```python
rfc=RandomForestClassifier(n_estimators=200,max_depth=None,random_state=42)
rfc.fit(X_train_scaled,y_train)
y_pred=rfc.predict(X_test_scaled)

result["Randon Forest"]=recall_score(y_test,y_pred)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(recall_score(y_test,y_pred))
```

# GRADIENT BOOSTING(ENSEMBLE LEARNING

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc=GradientBoostingClassifier(
n_estimators=200,
random_state=42
)

gbc.fit(X_train_scaled,y_train)
y_pred=gbc.predict(X_test_scaled)

result["Gradient Boosting"]=recall_score(y_test,y_pred)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(recall_score(y_test,y_pred))
```

# VOTING CLASSIFIER {ensemble learning}

```python

vc=VotingClassifier(
    estimators=[
        ("lr",LogisticRegression(penalty="l2",solver="liblinear",max_iter=1000)),
        ("knn",KNeighborsClassifier(n_neighbors=5,metric="euclidean")),
        ("rf",RandomForestClassifier(n_estimators=200,max_depth=None,random_state=42))
    ],
    voting='hard'
)
vc.fit(X_train_scaled,y_train)
y_pred=vc.predict(X_test_scaled)

result["Voting Classifier"]=recall_score(y_test,y_pred)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(recall_score(y_test,y_pred))
```
