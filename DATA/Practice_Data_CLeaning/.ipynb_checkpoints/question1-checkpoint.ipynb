```python
import json

def dumper(filename):
    with open(filename, "r") as f:
        data1=json.load(f)
    print("Data successfully fetched\n ", data1)
    return data1
    
        
```

```python
data=dumper("question1.json")
```

```python
def cleaning(data):
    ages=data["data"]["ages"]
    clean_ages=[]
    for i in ages:
        if i is None:
            continue
        if isinstance(i,str):
            i=i.strip()
            if not i.isdigit():
                continue
            i=int(i)
        if i<=0 or i>120:
            continue
        
        clean_ages.append(i)
    return print(clean_ages)
           
             
cleaning(data)
```
