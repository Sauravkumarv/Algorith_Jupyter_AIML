```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)
    return data
       
```

```python
data=load_data("store_data.json")
print(data)
print(type(data))
```

```python
# clean and structure the data
def cleaned_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    clean_data=[]
    unique_data=set()
    for user in data:
        raw_rating=str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating=text_to_num[raw_rating]
            user["rating"]=raw_rating
        #Now filled missing age with none
        raw_age=user.get("age")
        if(raw_age==None):
            user["age"]=None
        #Duplication of name
        if(user["name"].strip() in unique_data):
            continue
        clean_data.append(user)
    
        
    return(clean_data)
```

```python
cl_data=cleaned_data(data)
```

```python
def insights(data):
    sum=0
    for user in data:
        sum+=float(user["rating"])
        avg_rating=sum/len(data)
    print(f"Average rating of users is: {avg_rating}")
    poor=0
    for user in data:
        if(float(user["rating"])<3):
            poor+=1
    print(f"% of poor ratings are: {poor/len(data)*100}")
```

```python
insights(cl_data)
```

```python
def recommandation(data):
     recommand=[]
     for user in data:
         curren_recom={}
         curren_recom["name"]=user["name"]
         if(float(user["rating"])>3):
             curren_recom["brand"]="Apple"
         else:
             curren_recom["brand"]="Samsung"
         recommand.append(curren_recom)
     print(recommand)
     
   
```

```python
recommandation(cl_data)
```
