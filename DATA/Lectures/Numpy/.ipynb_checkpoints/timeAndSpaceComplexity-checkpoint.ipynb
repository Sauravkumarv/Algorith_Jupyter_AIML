```python
import numpy as np
import time
size=100_000_000
py_list=list(range(size))
start=time.time()
cal_sq=(a**2 for a in py_list)
end=time.time()

print(f"Python List time={end-start} seconds")

py_arr=np.array(py_list)
start=time.time()
cal_sq=py_arr**2   #Vectorization
end=time.time()

print(f"Python Array time={end-start} seconds")



```
