# Creating Arrays Using Built In Method

- ( Prefill ) Zero Matrix

```python
import numpy as np
arr1=np.zeros((2,3))
print(f"{arr1}{arr1.shape}")
```

- ( Prefill ) Ones Matrix

```python
arr2=np.ones((5,))
print(f"{arr2}{arr2.shape}")
```

-  Prefill With Value

```python
arr3=np.full((3,5),100)
print(f"{arr3}{arr3.shape}")
```

- Indentity Matrix

```python
arr4=np.eye(3)
print(f"{arr4}{arr4.shape}")
```

- Arrange Identity Matrix

```python
arr5=np.arange(0,11,2)       # work like range
print(f"{arr5}{arr5.shape}")
```

- Evenly Space Matrix

```python
arr6 = np.linspace(0,100,5,dtype="int64")     # line space work covers 100 in 5 steps
print(arr6,arr6.shape)
```
