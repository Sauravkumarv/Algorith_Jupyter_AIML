# Operations On Arrays

- Reshapping Arrays

```python
import numpy as np
arr=np.array([[1,4,5],[8,9,5]])

print(f"Before Reshape: \n{arr}\n{arr.shape}")

print(f"After Reshape: \n{arr.reshape(3,2)}")
```

```python
arr_flatten=arr.flatten()
print(f"{arr_flatten} \n{arr_flatten.shape}")
```

# Indexing Arrays

 - Normal Indexing

```python
arr1=np.array([[1,4,5],[8,9,7]])
print(arr1[1])
print(arr1[1][0])
```

- Fancy Indexing

```python
idx=[1]
print(arr[idx])
```

- Boolean Indexing

```python
arr2=np.array([7,8,5,4])
print(arr2[arr2%2==0])
```

- Slicing In Index  ( In Numpy while slicing Internally create view and in  list cretas a copy of the slicing array ) that's  the reason numpy is faster than list

```python
print(arr2[::2])
```
