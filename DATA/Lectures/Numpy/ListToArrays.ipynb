# List To Array

```python
import numpy as np
arr=np.array([1,2,3,4,5])
print(arr,type(arr),arr.dtype) # dtype tell us what type of array it is like it is of type Intefer (int64)
print(arr,type(arr),arr.shape) # tell us the size of the array


# Homogenious in nature 

arr2=np.array([1,2,3,4,5,"numpy"])
print(arr2,type(arr2),arr2.dtype) # dtype tell us what type of array it is like it is of type string (U21)
print(arr2,type(arr2),arr.shape)  # tell us the size of the array
```

```python
arr3=np.array([[7,8,9],[4,5,6],[1,2,3]])
print(f"{arr3} \n{type(arr3)}\n{arr3.dtype}\n{arr3.shape}") 
```
